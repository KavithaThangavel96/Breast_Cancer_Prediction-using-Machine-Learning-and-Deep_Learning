```python
#Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["font.size"]=14
plt.rcParams["figure.figsize"]=(10,6)
plt.rcParams["figure.facecolor"]="#D3D3D3"
```

```python
#load the data
path="breast_cancer.csv"

df=pd.read_csv(path)
```

```python
#top 5 data

df.head()
```

```python
df.drop(["id","Unnamed: 32"],axis=1,inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df["diagnosis"].value_counts()

#B-Benign_tumor(non-cancerous)-0
#M-Malignant_tumor(cancerous)-1
```

```python
#distribution of target variable

sns.countplot(x=df["diagnosis"],data=df)
```

```python
def encode(d):
    return d.map({"B":1,"M":0})
    
```

```python
df["diagnosis"]=df[["diagnosis"]].apply(encode)
```

```python
df.tail()
```

```python
#correlation

sns.heatmap(df.iloc[:,:10].corr(),annot=True)
```

```python
#pair plot
sns.pairplot(df.iloc[:, :10], hue="diagnosis")
```

```python
# split the model as train,test

x=df.drop(["diagnosis"],axis=1)
y=df["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train_scaled= scaler.fit_transform(x_train)

x_test_scaled=scaler.transform(x_test)
```

# Model Training

## Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
model=LogisticRegression()
model.fit(x_train_scaled,y_train)
```

```python
#traning accuracy and confusion matrix

y_train_predict=model.predict(x_train_scaled)
accuracy_score(y_train,y_train_predict)
```

```python
#testing accuracy and confusion matrix

y_test_predict=model.predict(x_test_scaled)
accuracy_score(y_test,y_test_predict)
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
Model_nb=GaussianNB()
Model_nb.fit(x_train_scaled,y_train)
```

```python
#traning accuracy and confusion matrix

y_train_predictnb=Model_nb.predict(x_train_scaled)
accuracy_score(y_train,y_train_predictnb)
```

```python
#testing accuracy and confusion matrix

y_test_predictnb=Model_nb.predict(x_test_scaled)
accuracy_score(y_test,y_test_predictnb)
```

# Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
Model_svc=SVC(kernel="linear")
Model_svc.fit(x_train_scaled,y_train)
```

```python
#traning accuracy and confusion matrix

y_train_predictsvc=Model_svc.predict(x_train_scaled)
accuracy_score(y_train,y_train_predictsvc)
```

```python
#testing accuracy and confusion matrix

y_test_predictsvc=Model_svc.predict(x_test_scaled)
accuracy_score(y_test,y_test_predictsvc)
```

## K Nearest Neighbhors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
score=[]

for k in range(1,20):
    model_knn=KNeighborsClassifier(n_neighbors=k)
    model_knn.fit(x_train_scaled,y_train)
    y_test_predictknn=model_knn.predict(x_test_scaled)
    score.append(accuracy_score(y_test,y_test_predictknn))
    
```

```python
score
```

```python
plt.plot([k for k in range(1,20)],score)
```

```python
model_knn=KNeighborsClassifier(n_neighbors=15)
model_knn.fit(x_train_scaled,y_train)
```

```python
#traning accuracy 
y_train_predictknn=model_knn.predict(x_train_scaled)
accuracy_score(y_train,y_train_predictknn)
```

```python
#testing accuracy
y_test_predictknn=model_knn.predict(x_test_scaled)
accuracy_score(y_test,y_test_predictknn)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dc=DecisionTreeClassifier()
model_dc.fit(x_train_scaled,y_train)
```

```python
#traning accuracy 
y_train_predictdc=model_dc.predict(x_train_scaled)
accuracy_score(y_train,y_train_predictdc)
```

```python
#testing accuracy
y_test_predictdc=model_dc.predict(x_test_scaled)
accuracy_score(y_test,y_test_predictdc)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf=RandomForestClassifier()
model_rf.fit(x_train_scaled,y_train)
```

```python
#traning accuracy 
y_train_predictrf=model_rf.predict(x_train_scaled)
accuracy_score(y_train,y_train_predictrf)
```

```python
#testing accuracy
y_test_predictrf=model_rf.predict(x_test_scaled)
accuracy_score(y_test,y_test_predictrf)
```
